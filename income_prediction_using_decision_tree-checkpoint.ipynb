```python
df = df[df['workclass']!= '?']
df = df[df['occupation']!= '?']
df = df[df['native.country']!= '?']
```

```python
df.head()
```

```python
df.categorical = df.select_dtypes(include =['object'])
df.categorical.apply(lambda x : x =='?').sum()
```

```python
df.rename(columns = {'marital.status': 'marital_status','native.country':'native_country','capital.gain':'capital_gain','capital.loss':'caital_loss','hours.per.week':'hours_per_week'},inplace = True)
```

```python
df.head()
```

```python
from sklearn import preprocessing
df_categorical=df.select_dtypes(include = ['object'])
```

```python
df_categorical.head()
```

```python
le = preprocessing.LabelEncoder()
```

```python
df_categorical = df_categorical.apply(le.fit_transform)
df_categorical.head()
```

```python
df= df.drop(df_categorical.columns, axis =1)
df= pd.concat([df,df_categorical], axis = 1)
```

```python
df.head()
```

```python
df.info()
```

```python
df['income'] = df['income'].astype('category')
```

```python
#importing train_test_select

from sklearn.model_selection import train_test_split

X = df.drop('income', axis =1)
y = df['income']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
X_train.shape
```

# importing decision tree classifier from sklearn

```python
from sklearn.tree import DecisionTreeClassifier

dt_default = DecisionTreeClassifier(max_depth = 5)
dt_default.fit(X_train, y_train)
```

# Let's check the evaluation metrics of our default model

# Importing classification report and confusion matrix from sklearn metrics

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
# Making predictions
y_pred_default = dt_default.predict(X_test)

# Printing classification report
print(classification_report(y_test, y_pred_default))
```

```python
# Printing confusion matrix and accuracy


print(confusion_matrix(y_test,y_pred_default))

print(accuracy_score(y_test,y_pred_default))
```
